```python
import pandas as pd

df = pd.read_csv("kariyer_ilanlari.csv")
df.head()
```

```python
# Şehir analizleri
print("Şehirlere göre ilan sayısı:")
print(df["Şehir"].value_counts())

# Firma analizleri
print("\nEn çok ilan veren firmalar:")
print(df["Firma"].value_counts())

# Pozisyon analizleri
print("\nEn çok geçen pozisyon başlıkları:")
print(df["Pozisyon"].value_counts())

# Tarih analizleri
df["Tarih"] = pd.to_datetime(df["Tarih"])
print("\nİlan tarihleri:")
print(df["Tarih"].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Şehir", order=df["Şehir"].value_counts().index, palette="Blues_d")
plt.title("Şehirlere Göre İlan Sayısı")
plt.xlabel("Şehir")
plt.ylabel("İlan Sayısı")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Firma"].value_counts().head(10).plot(kind="barh", color="orange")
plt.title("En Çok İlan Veren İlk 10 Firma")
plt.xlabel("İlan Sayısı")
plt.ylabel("Firma")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,7))
df["Pozisyon"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Pozisyon Dağılımı")
plt.ylabel("")  # pasta grafikte y ekseni olmamalı
plt.show()
```

```python
df["Tarih"] = pd.to_datetime(df["Tarih"])

plt.figure(figsize=(8,4))
df["Tarih"].value_counts().sort_index().plot(marker='o', linestyle='-')
plt.title("İlanların Yayın Tarihleri")
plt.xlabel("Tarih")
plt.ylabel("İlan Sayısı")
plt.grid(True)
plt.tight_layout()
plt.show()
```
